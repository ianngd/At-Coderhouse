```python
import pandas as pd
import requests
from plyer import notification
from IPython.display import display


def alerta (nivel,base,etapa):
    """Exibe notificação em caso de falha de carregamento de base"""
    notification.notify(
        title = 'Alerta Baixo',
        message = 'Falha do carregamento da base {} na etapa de {}'.format(base,etapa),
        app_name = 'Nome do aplicativo', timeout = 10
        )

url = "https://brasilapi.com.br/api/ibge/uf/v1"
response = requests.get(url)

if response.status_code == 200:
    data_json = response.json ()    
else:
    alerta(1,'IBGE','Extração')

estados_t = pd.DataFrame(data_json)

regiao = [estados_t['regiao']['nome'] for estados_t in data_json]
nome = [estados_t['nome'] for estados_t in data_json]
sigla = [estados_t['sigla'] for estados_t in data_json]


total_simples = pd.DataFrame({
    'Nome': nome,
    'Sigla': sigla,
    'Regiao': regiao
})

print ('Unidades federativas: ',estados_t['nome'].count())
 
regiao1 = 'Sul'
estados_sul = total_simples[total_simples['Regiao'] == regiao1]
estados_sul = estados_sul.sort_values(by='Nome')


regiao2 = 'Centro-Oeste'
estados_co = total_simples[total_simples['Regiao'] == regiao2]
estados_co = estados_co.sort_values(by='Nome')


regiao3 = 'Sudeste'
estados_su = total_simples[total_simples['Regiao'] == regiao3]
estados_su = estados_su.sort_values(by='Nome')


#Tabela 1:
display(estados_co.head())

#Tabela 2:
display(estados_su.head())

#Tabela 3:
display(estados_sul.head())
```
